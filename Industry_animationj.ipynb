```python
def nice_axes(ax):
    ax.set_facecolor('.8')
    ax.tick_params(labelsize=6, length=0)
    ax.grid(True, axis='x', color='white')
    ax.set_axisbelow(True)
    [spine.set_visible(False) for spine in ax.spines.values()]
    
nice_axes(ax)
fig
```

```python
dates = list(df_new.index)
dates
```

```python
fig, ax_array = plt.subplots(nrows=1, ncols=len(dates), figsize=(7, 2.5), dpi=144, tight_layout=True)
#dates = ["2020-03-07"]
for ax, date in zip(ax_array, dates):
    s = df_new.loc[date].sort_values()
#     y = df_new.loc[date].rank(method='first').values
    ax.barh(y=s.index, width=s.values, color=colors, tick_label=s.index)
    ax.set_title(dates, fontsize='smaller')
    nice_axes(ax)
    
```

```python
#Transition smoother
df2 = df_new.reset_index()
```

```python
df2.index = df2.index * 10
df2
```

```python
# fill in spaces between index values in order to extend animation
last_idx = df2.index[-1] + 1
df_expanded = df2.reindex(range(last_idx))
df_expanded
```

```python
df_expanded['index'] = df_expanded['index'].fillna(method='ffill')
df_expanded = df_expanded.set_index('index')
df_expanded
```

```python
#interpolate for clean animation transitions
df_expanded = df_expanded.astype(float).interpolate()
df_expanded
```

```python
#rank values so they will reorder during animation
df_rank_expanded = df_expanded.astype(float).rank(axis=1, method='first')
df_rank_expanded
```

```python
#plot each step of transition
fig, ax_array = plt.subplots(nrows=1, ncols=6, figsize=(12, 2), 
                             dpi=144, tight_layout=True)
labels = df_expanded.columns
for i, ax in enumerate(ax_array.flatten()):
    y = df_rank_expanded.iloc[i]
    width = df_expanded.iloc[i]
    ax.barh(y=y, width=width, color=colors, tick_label=labels)
    nice_axes(ax)
ax_array[0].set_title('2020-03-07')
ax_array[-1].set_title('2020-03-14');
```

```python
#create animation
from matplotlib.animation import FuncAnimation

def init():
    ax.clear()
    nice_axes(ax)
    ax.set_ylim(0, 10)

    pos1 = ax.get_position() # get the original position 
    pos2 = [pos1.x0+.3, pos1.y0,  pos1.width / 2.0, pos1.height / 2.0] 
    ax.set_position(pos2)
def update(i):
    for bar in ax.containers:
        bar.remove()
    y = df_rank_expanded.iloc[i]
    width = df_expanded.iloc[i]
    ax.barh(y=y, width=width, color=colors, tick_label=labels)
    date_str = df_expanded.index[i].strftime('%B %d, %Y')
    ax.set_title(f'Texas Unemployment Claims by Industry - {date_str}', fontsize='smaller')
    ax.set_xlim(0,df_expanded.iloc[i].max()+1000)
    
fig = plt.Figure(figsize=(8, 5), dpi=144)

ax = fig.add_subplot()
anim = FuncAnimation(fig=fig, func=update, init_func=init, frames=len(df_expanded), 
                     interval=100, repeat=False)
```
